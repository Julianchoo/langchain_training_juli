### prompt templates
1) Come up with a template, leaving a placeholder for an input variable
2) Create a prompt by calling PromptTemplate(), passing input variables and the template you created in 1)
3) Create a final prompt where the input variable takes a specific value. This is done by applying the .format method to your prompt in 2, and making the input variable you want to change equal to the value 
4) call llm() passing the final prompt

```python
my_template = """
I really want to travel to {location}, what should I do there?
"""


prompt = PromptTemplate(
    input_variables = ['location'],
    template=my_template,
)

final_prompt = prompt.format(location='Rome')

print(llm(final_prompt))
```

## Chains: looking at langchains documentation once more
A chain in LangChain is made up of links, which can be either primitives like LLMs or other chains.

The most core type of chain is an LLMChain, which consists of a PromptTemplate and an LLM.

Extending the previous example, we can construct an LLMChain which takes user input, formats it with a PromptTemplate, and then passes the formatted response to an LLM.

```python
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI

llm = OpenAI(temperature=0.9, openai_api_key = my_api_key, max_tokens=100)

prompt=PromptTemplate(
    input_variables=['product'], 
    template="give 3 strategic focus areas for a company that sells {product}"
)

```

```python
# This chain will take a user input, use it to format the prompt, and send it to the llm
from langchain.chains import LLMChain

chain = LLMChain(llm=llm, prompt=prompt)

chain.run("flying cars")
```

```python
# New chain, with a prompt that takes several inputs
from langchain.chains import LLMChain

prompt = PromptTemplate(
    input_variables=['product','demographic_segment','brand_style'],
    template = """
    come up with an SEM ad tagline for {product}, aimed at people in the following market: <{demographic_segment}>. The brand is considered <{brand_style}>
    """,
)

user_product = input("What is your product? Answer: ")
user_demographic_segment = input("Who is your target customer? Answer: ")
user_brand_style = input("What is your brand's style? Answer: ")

# This is how we usually did it
# --> final_prompt = prompt.format(product=user_product, demographic_segment=user_demographic_segment, brand_style=user_brand_style)

#Now with chains
chain = LLMChain(llm=llm,prompt=prompt)

# Create dictionary of input variables and values
user_inputs={'product':user_product, 'demographic_segment':user_demographic_segment,'brand_style':user_brand_style}
print(user_inputs)

response = chain.run(user_inputs)
print(response)
```

### Simple sequential chains from the guy in the video

```python
from langchain.llms import OpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.chains import SimpleSequentialChain

llm= OpenAI(temperature=0,openai_api_key=my_api_key)
```

```python
idea_template = """
Come up with an ad for a business that the user suggests.
% USER'S IDEA
{user_idea}

YOUR RESPONSE:
"""

prompt_template = PromptTemplate(input_variables=['user_idea'],template=idea_template)

idea_chain = LLMChain(llm=llm, prompt=prompt_template)
```

```python
ad_template = """
Given an ad, come up with a company name
% suggested ad: {ad}

YOUR RESPONSE:
"""

prompt_template = PromptTemplate(input_variables=['ad'],template=ad_template)

ad_chain = LLMChain(llm=llm, prompt=prompt_template)
```

```python
overall_chain = SimpleSequentialChain(chains=[idea_chain,ad_chain], verbose = True)
```

```python
a = overall_chain.run("Condoms for cats")
```
